# Cost Function for Linear regression


** Root Mean Squared Error **

$$
RMSE(\mathbf{X},h)= \sqrt{\frac{1}{m} \sum_{i=1}^{m}(h(\mathbf{x}_{i}) - y_{i})^2}
$$
